```python
import random
# --------------------------------------------------------- codigo de DG --------------------------------------------------------------------
def valiEdad(edad):
    if edad >= 18:
        return True
    else:
        print ("_______________________________________________________________________________________________")
        print("Mi estimado apostador debes tener al menos 18 años para jugar.")
        return False
# --------------------------------------------------------- codigo de DG --------------------------------------------------------------------
def valiTarj():
    while True:
        print ("_______________________________________________________________________________________________")
        tarjeta = input("Introduce el número de tu tarjeta, no te vamos a robar (solo números, 10 dígitos): ")
        if tarjeta.isdigit() and len(tarjeta) == 10:
            return True
        else:
            print ("_______________________________________________________________________________________________")
            print("Número de tarjeta inválido. Debe tener 10 dígitos.")
# ---------------------------------------------------------- codigo de DG ----------------------------------------------------------------------
def recarga():
    print ("_______________________________________________________________________________________________")
    recargas = [10, 30, 50, 100]
    print("Elige una cantidad para recargar: 10, 30, 50, 100")
    while True:
        try:
            print ("_______________________________________________________________________________________________")
            recarga = int(input("Cuanto deseas recargar para apostar:  "))
            if recarga in recargas:
                return recarga
            else:
                print ("_______________________________________________________________________________________________")
                print("Por favor elige una cantidad válida.")
        except ValueError:
            print ("_______________________________________________________________________________________________")
            print("Debe ser un número entero.")
print ("_______________________________________________________________________________________________")
equipos = [
    "Arsenal", "Aston Villa", "Bournemouth", "Brentford", "Brighton", "Chelsea",
    "Crystal Palace", "Everton", "Fulham", "Ipswich Town", "Leicester City",
    "Liverpool", "Manchester City", "Manchester United", "Newcastle United",
    "Nottingham Forest", "Southampton", "Tottenham Hotspur", "West Ham United",
    "Wolverhampton"
]
print ("_______________________________________________________________________________________________")
# ---------------------------------------------------------------- codigo de AM -------------------------------------------------------------------
def generar(equipos):
    partidos = []
    random.shuffle(equipos)
    for i in range(0, len(equipos) - 1, 2):
        equipol = equipos[i]
        equipov = equipos[i + 1]
        golesl = random.randint(0, 5)
        golesv = random.randint(0, 5)
        partidos.append((equipol, equipov, golesl, golesv))
    return partidos
# ------------------------------------------------------------------ codigo de DG ------------------------------------------------------------------
texto = """
_________________________________________________________________________________________________________________
||                                                                                                             ||
||           ███████╗ █████╗           ███████╗██╗   ██╗████████╗ ██████╗██████╗  █████╗ ██████╗ ████████╗     ||
||            ██╔════╝██╔══██╗         ██╔════╝██║   ██║╚══██╔══╝██╔════╝██╔══██╗██╔══██╗██╔══██╗╚══██╔══╝     ||
||            █████╗  ███████║ █████╗  █████╗  ██║   ██║   ██║   ╚█████╗ ██████╔╝██║  ██║██████╔╝   ██║        ||
||            ██╔══╝  ██╔══██║ ╚════╝  ██╔══╝  ██║   ██║   ██║    ╚═══██╗██╔═══╝ ██║  ██║██╔══██╗   ██║        ||
||            ██║     ██║  ██║         ██║     ╚██████╔╝   ██║   ██████╔╝██║     ╚█████╔╝██║  ██║   ██║        ||
||            ╚═╝     ╚═╝  ╚═╝         ╚═╝      ╚═════╝    ╚═╝   ╚═════╝ ╚═╝      ╚════╝ ╚═╝  ╚═╝   ╚═╝        ||
||                                                                                                             ||
||                        █████╗ ██████╗ ██╗   ██╗███████╗ ██████╗████████╗ █████╗  ██████╗                    ||
||                       ██╔══██╗██╔══██╗██║   ██║██╔════╝██╔════╝╚══██╔══╝██╔══██╗██╔════╝                    ||
||                       ███████║██████╔╝██║   ██║█████╗  ╚█████╗    ██║   ███████║╚█████╗                     ||
||                       ██╔══██║██╔═══╝ ██║   ██║██╔══╝   ╚═══██╗   ██║   ██╔══██║ ╚═══██╗                    ||
||                       ██║  ██║██║     ╚██████╔╝███████╗██████╔╝   ██║   ██║  ██║██████╔╝                    ||
||                       ╚═╝  ╚═╝╚═╝      ╚═════╝ ╚══════╝╚═════╝    ╚═╝   ╚═╝  ╚═╝╚═════╝                     ||
_________________________________________________________________________________________________________________
"""
print (texto)

def juego():
    print ("_______________________________________________________________________________________________")
    print("Bienvenido a FA FUTSPORT APUESTAS")
    print ("_______________________________________________________________________________________________")
    nombre = input("Ingresa tu nombre: ")
    try:
        print ("_______________________________________________________________________________________________")
        edad = int(input("Ingresa tu edad: "))
    except ValueError:
        print ("_______________________________________________________________________________________________")
        print("La edad debe ser un número.")
        return

    if not valiEdad(edad):
        return

    if not valiTarj():
        return
# --------------------------------------------------------- codigo de AM --------------------------------------------------------------------
    saldo = recarga()
    tapostado = 0
    tganado = 0

    partidos = generar(equipos[:])
# --------------------------------------------------------- codigo de AM --------------------------------------------------------------------
    while True:
        print ("_______________________________________________________________________________________________")
        print("\n--- Partidos disponibles ---")
        for i, (equipol, equipov, golesl, golesv) in enumerate(partidos):
            print(f"{i+1}. {equipol} vs {equipov}")

        while True:
            try:
                print ("_______________________________________________________________________________________________")
                eleccion = int(input(f"Elige el número del partido (1-{len(partidos)}): "))
                if 1 <= eleccion <= len(partidos):
                    partido = partidos.pop(eleccion - 1)
                    break
                else:
                    print ("_______________________________________________________________________________________________")
                    print(f"Por favor elige un número entre 1 y {len(partidos)}.")
            except ValueError:
                print ("_______________________________________________________________________________________________")
                print("Debes elegir un número válido.")

        equipol, equipov, golesl, golesv = partido

        while True:
            try:
                print ("_______________________________________________________________________________________________")
                apuesta = int(input(f"¿Cuánto deseas apostar a {equipol} vs {equipov}? (Saldo disponible: {saldo}): "))
                if apuesta > saldo:
                    print ("_______________________________________________________________________________________________")
                    print("No tienes suficiente saldo para realizar esta apuesta.")
                elif apuesta <= 0:
                    print ("_______________________________________________________________________________________________")
                    print("La apuesta debe ser mayor a 0.")
                else:
                    saldo -= apuesta
                    tapostado += apuesta
                    break
            except ValueError:
                print ("_______________________________________________________________________________________________")
                print("Debes ingresar un número válido para la apuesta.")

        while True:
            print ("_______________________________________________________________________________________________")
            prediccion = input(f" Cual es tu predicción de goles para {equipol} vs {equipov} (ej. 2-1): ")
            try:
                predlocal, predvisitante = map(int, prediccion.split('-'))
                break
            except ValueError:
                print("La predicción debe ser en el formato 'goles local-goles visitante'.")

        if predlocal == golesl and predvisitante == golesv:
            print ("_______________________________________________________________________________________________")
            print(f"¡FELICIADES! ¡Has ganado! El resultado fue {golesl}-{golesv}.")
            saldo += apuesta * 2
            tganado += apuesta * 2
        else:
            print(f"Ni modo rey perdite tu dinero. El resultado fue {golesl}-{golesv}.")
            print ("_______________________________________________________________________________________________")
        if len(partidos) == 0:
            print("No quedan más partidos disponibles.")
            break
        print ("_______________________________________________________________________________________________")
        continuar = input("¿Quieres seguir perdiento tu dinero digo digo apostando? (si/no): ")
        print ("_______________________________________________________________________________________________")
        if continuar.lower() != 'si':
            print(f"\nJuego terminado. Total apostado: {tapostado}, Total ganado: {tganado}, Saldo final: {saldo}")
            print ("_______________________________________________________________________________________________")
            break
if __name__ == "__main__":
    juego()
```
